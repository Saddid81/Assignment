```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets 
```

```python
# loading the Iris dataset
iris= datasets.load_iris()
type(iris)
```

```python
# Checking dataset caracteristics
iris.feature_names
```

```python
x= iris.data
y= iris.target
```

```python
# Dataset descomposition using Principal component analysis (PCA) -> https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
pca = decomposition.PCA(n_components=3) # n_components=3 because is for a 3D chart
pca.fit(x)
X = pca.transform(x)
```

```python
figura = plt.figure(1, figsize=(8, 7))
ax = Axes3D(figura,elev=45, azim=145)
for label, target in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(np.mean(X[y == target, 0]),
              np.mean(X[y == target, 1]),
              np.mean(X[y == target, 2]), label, fontsize=15, color='r', fontweight=100,
              horizontalalignment='center',
             bbox=dict(alpha=.55, edgecolor='w', facecolor='w'))
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.RdYlGn, edgecolor='k', s=100)
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
